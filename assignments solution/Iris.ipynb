```
df=pd.read_csv("Iris.csv")
```

# describing about data

```
df.describe()
```

```
df.describe().plot(kind = "barh",fontsize=15, figsize = (15,10), table = False,colormap="rainbow")
plt.xlabel('Value')
plt.ylabel('Statistics')
plt.title("General Statistics of Iris Dataset")
```

# bar plot for count of species in visualization

```
sns.countplot('Species',data=df)
#ax.set_title('Iris Species Count')
plt.show()
```

```
df['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```

```
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=df, kind="reg")
```

```
sns.pairplot(data=df, hue='Species',kind='reg')
```

```
sns.pairplot(data=df, hue='Species',kind='scatter')
```
