# Visualization 2 (seaborn)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
'''
seaborn provides better plots as compared to matplotlib aslo provide some automated calculation 
so seaborn used by all ML and Data processing Engineers 
'''
```

```python
df=pd.read_csv("data/Automobile_data.csv")
```

```python
df
```

# 1) Line Plot 

```python
x=np.arange(10)
y=np.sqrt(x)
```

```python
plt.figure(figsize=(6,6))
sns.lineplot(x,y)
plt.show()
```

# 2) Histogram (Distplot)

```python
plt.figure(figsize=(6,6))
sns.distplot(df["price"])
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(6,6))
sns.distplot(df["width"])
plt.show()
```

```python
width_mean=df["width"].mean()
width_median=df["width"].median()
```

```python
plt.figure(figsize=(6,6))
sns.distplot(df["width"])
plt.axvline(width_mean,color='red')
plt.axvline(width_median,color='yellow')
plt.show()
```

# 3) Count Plot (Bar plot)

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df,x="body-style")
#uplt.show()
```

```python
# how to seperate all categorical column in one data frame 
# once we done seperation we wants to use loop to represnt countplot of each 
plt.figure(figsize=(6,6))
sns.countplot(data=df,x="make")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df,y="make")
plt.show()
```

```python
'''in matplotlib we have legend here we have hue '''
plt.figure(figsize=(6,6))
sns.countplot(data=df,x="body-style",hue="fuel-type")
plt.show()
```

# 4) Heatmap 

```python
'''helps us to check correlation between features it best way to see correlation of 
features when we have more than 100 features '''
```

```python
corr=df.corr()

plt.figure(figsize=(7,7))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(corr,annot=True,cmap="viridis")
plt.show()
```

# 5) scatter plot

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df,x="width",y="price")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df,x="width",y="price",hue="body-style")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df,x="width",y="price",hue="make")
plt.show()
```

# 6) Box Plots 

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df,x="price")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df, x="price",y="make")
plt.show()
```
